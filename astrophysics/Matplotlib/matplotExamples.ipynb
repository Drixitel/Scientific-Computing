```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.arange(0,5,0.1)   #create x=[0,5) in 0.1 increemnts 
y = np.sin(x)            # y = sin(x)

plt.plot(x)        # make a plot
plt.ylabel('sin(x)')     # label y axis 
plt.xlabel('x')          # label x axis 


# save the file 
# plt.savefig('sinx.png', bbox_inches = 'tight', dpi=300)
```

```python
x = np.linspace(0,2*np.pi,100) 
y = np.sin(x)          #these are 4 arrays that take the x array
z = np.cos(x)
w = np.sin(4*x)
v = np.cos(4*x)

#call subplots functions to generate a multiplot figure


f, axarr = plt.subplots(1,2)

#treat axarr as an array, from left to right

axarr[0].plot(x,y)
axarr[0].set_xlabel('x') 
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

#second pannel 

axarr[1].plot(x,z)
axarr[1].set_xlabel('x')  
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

#whole figure adj. using f 
#add more space between the figures 

f.subplots_adjust(wspace=.5) #spaced out the plot

#wspace is reserved to the width between the subplots 
```

```python
#Testing to see if we can use arb. names 
#result we can 

no, yes = plt.subplots(1,2)

#first panel

yes[0].plot(x,y)
yes[0].set_xlabel('x') 
yes[0].set_ylabel('sin(x)')
yes[0].set_title(r'$\sin(x)$')

#second pannel 

yes[1].plot(x,z)
yes[1].set_xlabel('x')  
yes[1].set_ylabel('cos(x)')
yes[1].set_title(r'$\cos(x)$')



no.subplots_adjust(wspace=0.4) 

#change individual subplots ratio
#set the axis ratio to be equal 

#fix the ratio
#here are two possible options 


#yes[0].set_aspect('equal') #set distance between tick marks to be equal 
#yes[1].set_aspect('equal') #we see this does not work because of the range

#range is from 3 ticks and domain is 4 ticks, they are euqal spaced 
#but not from each other; does not look good 


#yes[0].set_aspect('auto') #is default does not work 


yes[0].set_aspect(np.pi) #ratio of y vs. x is set to pi 
```

```python
#adjust the size of the figure 
fig = plt.figure(figsize = (6,6))

plt.plot(x, y, label= r'$y = \sin(x)$')    #add a label to the line
plt.plot(x, z, label= r'$y = \cos(x)$')    #add a label to the secod line
plt.plot(x, w, label= r'$y = \sin(4x)$')   #add a label to the third line
plt.plot(x, v, label= r'$y = \cos(4x)$')   #add a label to the fourth line


#see notes about how to change the color of a line 
#plt.plot(x, v, color = 'purple', label= r'$y= \cos(4x)$')

plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')

plt.xlim([0,2*np.pi])                      #set the bounds of x axis 
plt.ylim([-1.2,1.2])                       #set the bounds of y axis 

#plt.xlim([,]) takes two elements the start and stop 

plt.legend(loc = 1,framealpha = 0.95)       #framealpha: transparency 




#fix the axis ratio 

plt.gca().set_aspect(np.pi/1.2)

# for linear plots: [deltax/ delta] y will give you the aspect value input 
```
