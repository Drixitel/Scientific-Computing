```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
```

```python
n = 10_000
```

```python
x = np.random.uniform(-1,1,n)
y = np.random.uniform(-1,1,n)
```

```python
plt.plot(x,y,'.', ms = 0.4)
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.gca().set_aspect('equal')
plt.show()
#using these points then enscribing the circle to find the area of the circle 
```

```python
ir = np.where((x**2 + y**2)<1.0) [0] #inner radius 
#np.where takes arrays as an inequality 
#where the condition is tru give the values 
#give me the index of that arraay 
#we obatin indecies 

ur = np.where((x**2 + y**2)>=1.0) [0]
```

```python
#plt.plot(x,y,'.', ms = 0.4)
#plt.xlim([-1,1])
#plt.ylim([-1,1])
#plt.gca().set_aspect('equal')
#plt.show()

fig = plt.figure(figsize =(7,7))

plt.xlim([-1.1,1.1])
plt.ylim([-1.1,1.1])
plt.plot(x[ir],y[ir], '.', color = 'blue')
plt.plot(x[ur],y[ur], '.', color = '0.75')
theta = np.linspace(0,2*np.pi, 1_000)
xc = np.cos(theta) #x coordinate 
yc = np.sin(theta)

plt.plot(xc,yc, color = 'green')#plot curve as line

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
pi_approx = 4.0* len(ir)/float(n)  #area of rectangle is 4 
#ratio len(inner)/ total samples 

error_pi = (pi_approx-np.pi)/np.pi 
print('Number of samples = ', n) 
print('Approximate pi = ', pi_approx)
print('Error in approx = ', error_pi)
```
