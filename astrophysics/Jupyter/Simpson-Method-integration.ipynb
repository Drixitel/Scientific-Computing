```python
# define the function 

def func(x): #f(x) takes a set of values 
    a = 1.01
    b = -3.04
    c = 2.07 
    return a*x**2 + b*x + c #returns a properformat 
```

```python
def func_integral(x): #F(x) : Used to check if our aproximation is correct 
    a = 1.01
    b = -3.04
    c = 2.07 
    return (a*x**3)/3 + (b*x**2)/2 + c*x
```

```python
def simpson_core(f,x,h):  #takes a function,and two values 
    
    return h*( f(x) + 4*f(x+h) + f(x+2*h))/3. 

#very very important note
#the core is already accounting for the values 
#1/3f(x_i) + 4/3f(x_(i+1)) + 1/3f(x_(i+2)) Mah fucking god  

#returns an interval evaluation 
```

# Answer



Simpson's method requires only 3 points and is deterministic of either even or odd intervals 



```python
def simpsons_method(f,a,b,N): 
    #f == function to integrate 
    #a == lowre limit of integration
    #b == upper limit of integration 
    #N == number of function evauations to use 
    # note if N is odd, then we don't need to adjust the
    # the last segment
    
    #define the x values to preform the simpsons rule 
    x = np.linspace(a,b,N)
    h = x[1]-x[0]          #same as Trapzoid, linspace is even, so just take first 2
    
    #define the value of the integral ..why?
    Fint = 0.0 
    
    #preform the integral using simpson's Method 
    for i in range(0,len(x)-2, 2): 
        Fint += simpson_core(f,x[i], h)      #this is an index call not a value call, the index is ran though a function
        
        
    #apply simpson's rule over the last interval
    #if N is even 
    if (N%2) == 0: #rule for even modulus 
        Fint += simpson_core(f,x[-2], 0.5*h)  #this accounts for the even intervals, need pairs of 3 
                                              # we would assume it would be an over estimate becuase we are taking 2 
                                            # values from the end, but you see the width is haved. 

    return Fint
```
