```python
%matplotlib inline
import matplotlib.pyplot as plt 
import matplotlib.colors as colors 
import numpy as np
```

```python
fname = 'density_projection.dat'
d = np.fromfile(fname, dtype =np.float32)
d = np.reshape(d, (512,512))
```

```python
fname = 'energy_projection.dat'
energy = np.fromfile(fname, dtype =np.float32)
energy = np.reshape(energy,(512,512))
```

```python
d_log = np.log10(d)
f = plt.figure(figsize = (10,10))
plt.imshow(d_log)
```

```python
e_log = np.log10(energy)
f = plt.figure(figsize =(10,10))
plt.imshow(e_log)
```

```python
d_min = d_log.min()
d_max = d_log.max()

v = (d_log - d_min)/ (d_max - d_min)
s = 1.0 - v
#s = v.copy()
#s[:] = 0.0
```

```python
e_min = e_log.min()
e_max = e_log.max()

h = 0.9 -0.3*(e_log - e_min)/(e_max - e_min)

#h = 0. -1.*(e_log - e_min)/(e_max - e_min)
```

```python
hsv_image = np.zeros((512,512,3))

hsv_image[:,:,0] = h 
hsv_image[:,:,1] = s 
hsv_image[:,:,2] = v 

rgb_image = colors.hsv_to_rgb(hsv_image)

```

```python
f = plt.figure(figsize = (10,10))
plt.imshow(rgb_image) 
plt.savefig('test_image.png', bbox_inches = 'tight', dpi=300)
```
